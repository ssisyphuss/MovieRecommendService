```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

print(stopwords.words('english'))
```

```python
# 정규식을 통해서 문자만 남김 
def sub_ex(s):
  return re.sub(r'[^ㄱ-ㅎㅏ-ㅣ가-힣0-9a-zA-Z ]','',s)


df_target.loc[:,'en_overview'] = df_target.loc[:,'en_overview'].apply(sub_ex)

N = len(df_target)

df_target.loc[:10, ['en_overview']]
print(df_target.loc[0, ['en_overview']][0])
```

```python
# nltk 라이브러리를 통해서 불용어 처리
stop_words = set(stopwords.words('english'))

def stop_words_fun(string):
    result = []
    word_tokens = word_tokenize(string)
    for w in word_tokens:
        if w not in stop_words:
            result.append(w)
    return ' '.join(result)

# print(stop_words_fun('computer is good'))


# 불용어 처리 후 모든 줄거리 컬럼의 내용을 처리된 결과로 바꾸어줌
for i in range(N):
    temp_string = df_target.loc[i, ['en_overview']][0]
    new_string = stop_words_fun(temp_string)
    df_target.loc[i, ['en_overview']] = new_string
    
    
df_target.loc[:10, ['en_overview']]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# TF-IDF 생성 후 수행
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df.en_overview)
```

```python
from sklearn.metrics.pairwise import linear_kernel

#  처리된 결과를 통해서 코사인 유사가 담긴 2차원 넘파이배열 추출
cosine_result = linear_kernel(tfidf_matrix, tfidf_matrix)
print(type(cosine_result))
```

```python
# 아이디를 통해서 인덱스를 접근하기 위해 사용
indices = pd.Series(df.index, index=df.movie_id)
# print(indices)
```

```python
m_copy = df.copy()
m_copy.head(10)
```

```python
from ast import literal_eval
```

```python
m_copy['genres'] = m_copy['genres'].apply(literal_eval)

m_copy['genres'] = m_copy['genres'].apply(lambda x : [ dic['genre_name'] for dic in x])

m_copy[['genres']]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# 리스트 객체를 문자열로 변환: 공백으로 구분
m_copy['genres_literal'] = m_copy['genres'].apply(lambda x : (' ').join(x))

# CountVectorizer
count_vect = CountVectorizer(min_df=0, ngram_range=(1,2))
genre_mat = count_vect.fit_transform(m_copy['genres_literal'])

print(genre_mat.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

genre_sim = cosine_similarity(genre_mat, genre_mat)
# print(type(genre_sim))
```

```python
id_list = list(m_copy.loc[:,'movie_id'])
print(id_list)
```

```python
def final_recommend(id):
    global id_list
    id_copy = id_list[:]
#     print(indices[id])
    id_copy.sort(key=lambda x: (0.9*cosine_result[indices[id]][indices[x]] + 0.1*genre_sim[indices[id]][indices[x]]), reverse=True)
    
    return id_copy[:10]
```

```python
final_recommend(11)
```

```python
recommendations = {}

for id in id_list:
    recommendations[id] = final_recommend(id)

print(recommendations)
```

```python
import pickle

with open('recommendations.p', 'wb') as file:
    recommendations = pickle.dump(recommendations, file)
```
